## 🏗️ Part 3: สร้าง Data Warehouse (40 นาที)

### Exercise 3.1: Health Profile Warehouse

**โจทย์**: สร้างตาราง Health Profile Warehouse ตามขั้นตอนต่อไปนี้:

**ขั้นตอน**:
1. ดึงข้อมูล**ล่าสุด**ของแต่ละสมาชิกจาก `ncds_profile` และ `ncds_record`
2. นับจำนวนครั้งการตรวจของแต่ละสมาชิก
3. JOIN กับ `campaign` เพื่อได้ `district_code`
4. JOIN กับ `districts_df` เพื่อได้ข้อมูลพื้นที่
5. สร้าง DataFrame ตามคอลัมน์ที่กำหนด

**คอลัมน์ที่ต้องมี**:
- member_id
- จำนวนครั้งการตรวจ
- เพศ (แปลง 0→หญิง, 1→ชาย)
- อายุ
- ตำบล_id, ตำบล
- อำเภอ_id, อำเภอ
- จังหวัด_id, จังหวัด
- เขต
- กลุ่มอายุ
- ประวัติการเจ็บป่วย (code และ label)
- ประวัติแพ้ยา (code และ label)
- ประวัติแพ้อาหาร (code และ label)
- ญาติสายตรงNCDs (code และ label)

```python
# Step 1: ดึงข้อมูลล่าสุดของแต่ละสมาชิก
# Hint: ใช้ .sort_values('dateadd').groupby('memberid').tail(1)

latest_profiles = ncds_profile.sort_values('dateadd').groupby('memberid').tail(1)
latest_records = ncds_record.sort_values('dateadd').groupby('memberid').tail(1)

print(f"✓ Latest profiles: {len(latest_profiles)} แถว")
print(f"✓ Latest records: {len(latest_records)} แถว")
```

```python
# Step 2: นับจำนวนครั้งการตรวจ
# Hint: ใช้ .groupby('memberid').size().reset_index(name='จำนวนครั้งการตรวจ')

screening_count = ncds_record.groupby('memberid').size().reset_index(name='จำนวนครั้งการตรวจ')

print(f"✓ Screening count: {len(screening_count)} สมาชิก")
screening_count.head()
```

```python
# Step 3: JOIN กับ campaign
# Hint: ใช้ pd.merge() โดย left_on='campaignId', right_on='id'

campaign['district_code'] = campaign['district_code'].astype(str)

profile_campaign = pd.merge(latest_profiles, campaign[['id', 'district_code']], left_on='campaignId', right_on='id', how='left')

print(f"✓ Profile + Campaign: {len(profile_campaign)} แถว")
```

```python
# Step 4: JOIN กับ districts
# Hint: ใช้ pd.merge() โดย on='district_code'

profile_location = pd.merge(profile_campaign, districts_df, on='district_code', how='left')

print(f"✓ Profile + Location: {len(profile_location)} แถว")
```

```python
# Step 5: JOIN กับ record และ screening_count
# Hint: JOIN 2 ครั้ง - ครั้งแรกกับ latest_records, ครั้งที่สองกับ screening_count

profile_record = pd.merge(profile_location, latest_records, on='memberid', how='left', suffixes=('_profile', '_record'))

final_data = pd.merge(profile_record, screening_count, on='memberid', how='left')

# แปลง NaN ในจำนวนครั้งการตรวจเป็น 0
final_data['จำนวนครั้งการตรวจ'] = final_data['จำนวนครั้งการตรวจ'].fillna(0).astype(int)

print(f"✓ Final data: {len(final_data)} แถว")
display(final_data.head())
```

```python
# Step 6: สร้าง Helper Functions
# TODO: เขียนฟังก์ชันแปลงค่าต่างๆ

def get_gender_label(gender):
    """แปลงเพศ: 0→หญิง, 1→ชาย, อื่นๆ→999"""
    if gender == 0:
        return 'หญิง'
    elif gender == 1:
        return 'ชาย'
    else:
        return 'ไม่ระบุ'

def get_age_group(age):
    """
    แปลงกลุ่มอายุ:
    - < 20: 'อายุต่ำกว่า 20 ปี'
    - 20-34: 'อายุระหว่าง 20-34 ปี'
    - 35-59: 'อายุระหว่าง 35-59 ปี'
    - >= 60: 'อายุ 60 ปีขึ้นไป'
    """
    if age < 20:
        return 'อายุต่ำกว่า 20 ปี'
    elif 20 <= age <= 34:
        return 'อายุระหว่าง 20-34 ปี'
    elif 35 <= age <= 59:
        return 'อายุระหว่าง 35-59 ปี'
    elif age >= 60:
        return 'อายุ 60 ปีขึ้นไป'
    else:
        return 'ไม่ระบุ'


def convert_illness_history(value):
    """
    แปลงประวัติการเจ็บป่วย:
    - -1: 'ไม่มีประวัติ'
    - 1: 'มีประวัติการเจ็บป่วยรวม'
    - อื่นๆ: 'ไม่ทราบ'
    """
    if value == -1:
        return 'ไม่มีประวัติ'
    elif value == 1:
        return 'มีประวัติการเจ็บป่วยรวม'
    else:
        return 'ไม่ทราบ'

# TODO: เขียนฟังก์ชันเพิ่มเติมสำหรับ drug_allergy, food_allergy, relatives_ncds
# (รูปแบบเหมือน convert_illness_history)

def convert_drug_allergy(value):
     """
     แปลงประวัติแพ้ยา:
     - -1: 'ไม่มีประวัติ'
     - 1: 'มีประวัติแพ้ยา'
     - อื่นๆ: 'ไม่ทราบ'
     """
     if value == -1:
         return 'ไม่มีประวัติ'
     elif value == 1:
         return 'มีประวัติแพ้ยา'
     else:
         return 'ไม่ทราบ'

def convert_food_allergy(value):
     """
     แปลงประวัติแพ้อาหาร:
     - -1: 'ไม่มีประวัติ'
     - 1: 'มีประวัติแพ้อาหาร'
     - อื่นๆ: 'ไม่ทราบ'
     """
     if value == -1:
         return 'ไม่มีประวัติ'
     elif value == 1:
         return 'มีประวัติแพ้อาหาร'
     else:
         return 'ไม่ทราบ'

def convert_relatives_ncds(value):
     """
     แปลงประวัติญาติสายตรงNCDs:
     - -1: 'ไม่มีประวัติ'
     - 1: 'มีประวัติญาติสายตรงNCDs'
     - อื่นๆ: 'ไม่ทราบ'
     """
     if value == -1:
         return 'ไม่มีประวัติ'
     elif value == 1:
         return 'มีประวัติญาติสายตรงNCDs'
     else:
         return 'ไม่ทราบ'
```

```python
# ดึงข้อมูลล่าสุดของแต่ละสมาชิก
latest_profiles = ncds_profile.sort_values('dateadd').groupby('memberid').tail(1)
latest_records = ncds_record.sort_values('dateadd').groupby('memberid').tail(1)
screening_count = ncds_record.groupby('memberid').size().reset_index(name='จำนวนครั้งการตรวจ')

# JOIN ข้อมูล
campaign['district_code'] = campaign['district_code'].astype(str)
profile_campaign = pd.merge(latest_profiles, campaign[['id', 'district_code']],
                            left_on='campaignId', right_on='id', how='left')
profile_location = pd.merge(profile_campaign, districts_df, on='district_code', how='left')
profile_record = pd.merge(profile_location,
                         latest_records[['memberid', 'drug_allergy', 'food_allergy', 'record_ncds', 'relatives_ncds']],
                         on='memberid', how='left')
final_data = pd.merge(profile_record, screening_count, on='memberid', how='left')
final_data['จำนวนครั้งการตรวจ'] = final_data['จำนวนครั้งการตรวจ'].fillna(0).astype(int)

# Helper functions
def handle_text_data(value):
    return 999 if pd.isna(value) or value == '' else str(value)

def handle_numeric_data(value):
    return 999 if pd.isna(value) else float(value)

def get_gender_label(gender):
    if pd.isna(gender): return 999
    return "หญิง" if int(gender) == 0 else "ชาย"

def get_age_group(age):
    if pd.isna(age) or age == 999 or age == 0: return 999
    if age < 20: return 'อายุต่ำกว่า 20 ปี'
    if age < 35: return 'อายุระหว่าง 20-34 ปี'
    if age < 60: return 'อายุระหว่าง 35-59 ปี'
    return 'อายุ 60 ปีขึ้นไป'

def convert_value(value, labels):
    """แปลงค่า -1, 1 เป็น label"""
    if pd.isna(value) or value == '' or value == 999: return labels[0]
    if value == -1: return labels[1]
    if value == 1: return labels[2]
    return labels[0]

# สร้าง Warehouse
health_profile_wh = pd.DataFrame({
    'member_id': final_data['memberid'],
    'จำนวนครั้งการตรวจ': final_data['จำนวนครั้งการตรวจ'],
    'เพศ': final_data['gender'].apply(get_gender_label),
    'อายุ': final_data['age'].apply(handle_numeric_data),
    'ตำบล_id': final_data['district_code'].apply(handle_text_data),
    'ตำบล': final_data['district_name'].apply(handle_text_data),
    'อำเภอ_id': final_data['amphur_code'].apply(handle_numeric_data),
    'อำเภอ': final_data['amphur_name'].apply(handle_text_data),
    'จังหวัด_id': final_data['province_code'].apply(handle_numeric_data),
    'จังหวัด': final_data['province_name'].apply(handle_text_data),
    'เขต': final_data['zone'].apply(handle_numeric_data),
    'กลุ่มอายุ': final_data['age'].apply(get_age_group),
    'ประวัติการเจ็บป่วย_code': final_data['record_ncds'].apply(handle_numeric_data),
    'ประวัติการเจ็บป่วย': final_data['record_ncds'].apply(lambda x: convert_value(x, ['ไม่ทราบ', 'ไม่มีประวัติ', 'มีประวัติการเจ็บป่วยรวม'])),
    'ประวัติแพ้ยา_code': final_data['drug_allergy'].apply(handle_numeric_data),
    'ประวัติแพ้ยา': final_data['drug_allergy'].apply(lambda x: convert_value(x, ['ไม่ทราบ', 'ไม่มีประวัติ', 'มีประวัติการแพ้ยา'])),
    'ประวัติแพ้อาหาร_code': final_data['food_allergy'].apply(handle_numeric_data),
    'ประวัติแพ้อาหาร': final_data['food_allergy'].apply(lambda x: convert_value(x, ['ไม่ทราบ', 'ไม่มีประวัติ', 'มีประวัติการแพ้อาหาร'])),
    'ญาติสายตรงNCDs_code': final_data['relatives_ncds'].apply(handle_numeric_data),
    'ญาติสายตรงNCDs': final_data['relatives_ncds'].apply(lambda x: convert_value(x, ['ไม่ทราบ', 'ไม่มีประวัติ', 'มีประวัติการเจ็บป่วยรวม']))
})

print(f"✅ Health Profile Warehouse: {len(health_profile_wh)} แถว")
health_profile_wh.head()
```

```python
# TODO: สร้าง Disease Risk Warehouse
# Hint: 1. ดึงค่าจาก memberdata (sys/dias, bmi, waistline, height)
#       2. สร้างฟังก์ชันคำนวณความเสี่ยงทั้ง 3 ประเภท
#       3. Loop แต่ละสมาชิกและคำนวณ
#       4. สร้าง DataFrame จากผลลัพธ์

# 1. ดึงค่าจาก memberdata
member_measurements = memberdata[memberdata['name'].isin(['sys/dias', 'bmi', 'waistline', 'height'])].copy()

# Pivot table to get measurements for each member in columns
member_measurements_pivot = member_measurements.pivot_table(
    index='memberid',
    columns='name',
    values='value',
    aggfunc='first'
).reset_index()

# Split sys/dias into sys and dias
if 'sys/dias' in member_measurements_pivot.columns:
    sys_dias = member_measurements_pivot['sys/dias'].str.split('/', expand=True)
    member_measurements_pivot['sys'] = pd.to_numeric(sys_dias[0], errors='coerce')
    member_measurements_pivot['dias'] = pd.to_numeric(sys_dias[1], errors='coerce')
    member_measurements_pivot = member_measurements_pivot.drop(columns=['sys/dias'])

# Convert other columns to numeric
for col in ['bmi', 'waistline', 'height']:
    if col in member_measurements_pivot.columns:
        member_measurements_pivot[col] = pd.to_numeric(member_measurements_pivot[col], errors='coerce')

# Merge with latest profile data for age, gender, relatives_diabetes
risk_data = pd.merge(member_measurements_pivot, latest_profiles[['memberid', 'age', 'gender']], on='memberid', how='left')
risk_data = pd.merge(risk_data, latest_records[['memberid', 'relatives_diabetes']], on='memberid', how='left')


# 2. สร้างฟังก์ชันคำนวณความเสี่ยงทั้ง 3 ประเภท

def calculate_hypertension_risk(sys, dias):
    """คำนวณความเสี่ยงความดันโลหิตสูง"""
    if pd.isna(sys) or pd.isna(dias):
        return 'ไม่ระบุ', 999
    if sys >= 140 or dias >= 90:
        return 'เสี่ยงสูง', 1
    elif sys >= 120 or dias >= 80:
         return 'เสี่ยงปานกลาง', 0.5
    else:
        return 'เสี่ยงต่ำ', 0

def calculate_findrisc_score(age, bmi, waistline, sys, relatives_diabetes):
    """คำนวณ FINDRISC score และระดับความเสี่ยง"""
    score = 0
    if pd.isna(age) or pd.isna(bmi) or pd.isna(waistline) or pd.isna(sys) or pd.isna(relatives_diabetes):
        return 'ไม่ระบุ', 999, 999

    # Age
    if age >= 45 and age <= 54: score += 2
    elif age >= 55 and age <= 64: score += 3
    elif age >= 65: score += 4

    # BMI
    if bmi >= 25 and bmi < 30: score += 1
    elif bmi >= 30: score += 3

    # Waistline (using female/male thresholds) - Assuming female if gender is 0, male if 1
    # We need gender for this, which is in latest_profiles. Let's assume we have it.
    # For simplicity here, let's use a generic threshold if gender is not available or unclear.
    # In a real scenario, you'd need to merge gender information.
    # Let's use generic thresholds for now or assume gender is available in risk_data
    # if gender == 0 (หญิง) and waistline >= 88: score += 4
    # elif gender == 1 (ชาย) and waistline >= 102: score += 4
    # elif gender == 0 (หญิง) and waistline >= 80 and waistline < 88: score += 3
    # elif gender == 1 (ชาย) and waistline >= 94 and waistline < 102: score += 3

    # Let's use a simplified approach for waistline based on available data structure (no gender directly here)
    # If you have gender, you'd modify this part. Assuming gender is available in risk_data now.
    # Need to merge gender from latest_profiles
    # Merged risk_data with latest_profiles already includes gender

    # Waistline based on gender
    gender = risk_data.loc[risk_data['memberid'] == memberid, 'gender'].iloc[0] if memberid in risk_data['memberid'].values else None

    if not pd.isna(gender):
        if int(gender) == 0: # หญิง
            if waistline >= 88: score += 4
            elif waistline >= 80: score += 3
        elif int(gender) == 1: # ชาย
            if waistline >= 102: score += 4
            elif waistline >= 94: score += 3
    # If gender is not available or invalid, skip waistline points or use a default logic if applicable

    # Sys (using the higher value if sys/dias exists)
    if sys >= 140: score += 2

    # Relatives with diabetes (assuming 1 means yes, -1 means no, 0 or NaN means unknown)
    if relatives_diabetes == 1: score += 5

    # Determine risk level
    if score < 7:
        risk_level = 'เสี่ยงต่ำ'
    elif score >= 7 and score <= 11:
        risk_level = 'เสี่ยงปานกลาง'
    elif score >= 12 and score <= 14:
        risk_level = 'เสี่ยงสูง'
    elif score >= 15 and score <= 20:
        risk_level = 'เสี่ยงสูงมาก'
    else: # score > 20 or invalid score
        risk_level = 'ไม่ระบุ'

    return risk_level, score, score

def calculate_cardiovascular_risk(age, gender, sys, waistline, height):
    """คำนวณความเสี่ยงโรคหัวใจและหลอดเลือด"""
    # This is a simplified calculation. Real cardiovascular risk scores (like Framingham, ASCVD)
    # are more complex and require more variables (cholesterol, blood pressure treatment, etc.).
    # Based on the prompt, we'll use the provided variables: age, gender, sys, waist/height ratio.

    risk_score = 0
    risk_level = 'ไม่ระบุ'

    if pd.isna(age) or pd.isna(gender) or pd.isna(sys) or pd.isna(waistline) or pd.isna(height) or height == 0:
         return risk_level, 999, 999

    # Age factor (example: higher age, higher risk)
    risk_score += age * 0.1

    # Gender factor (example: male higher risk)
    if int(gender) == 1: risk_score += 5

    # Systolic Blood Pressure factor
    risk_score += sys * 0.1

    # Waist-to-Height Ratio
    waist_height_ratio = waistline / height if height > 0 else 0
    risk_score += waist_height_ratio * 20 # Example factor

    # Determine risk level based on score (example thresholds)
    if risk_score < 20:
        risk_level = 'เสี่ยงต่ำ'
    elif risk_score < 40:
        risk_level = 'เสี่ยงปานกลาง'
    else:
        risk_level = 'เสี่ยงสูง'

    return risk_level, risk_score, risk_score


# 3. Loop แต่ละสมาชิกและคำนวณ & 4. สร้าง DataFrame จากผลลัพธ์
disease_risk_list = []

# Ensure all necessary columns are in risk_data, handle potential missing columns gracefully
required_cols = ['memberid', 'age', 'gender', 'bmi', 'waistline', 'height', 'sys', 'dias', 'relatives_diabetes']
for col in required_cols:
    if col not in risk_data.columns:
        risk_data[col] = np.nan # Add missing columns with NaN

for index, row in risk_data.iterrows():
    memberid = row['memberid']
    age = row['age']
    gender = row['gender']
    bmi = row['bmi']
    waistline = row['waistline']
    height = row['height']
    sys = row['sys']
    dias = row['dias']
    relatives_diabetes = row['relatives_diabetes'] # This comes from latest_records



    # Calculate Hypertension Risk
    hypertension_level, hypertension_score = calculate_hypertension_risk(sys, dias)
    disease_risk_list.append({
        'member_id': memberid,
        'ภาวะเสี่ยง': 'โรคความดันโลหิตสูง',
        'ระดับความเสี่ยง': hypertension_level,
        'คะแนน': hypertension_score
    })

    # Calculate FINDRISC
    findrisc_level, findrisc_score, findrisc_raw_score = calculate_findrisc_score(age, bmi, waistline, sys, relatives_diabetes)
    disease_risk_list.append({
        'member_id': memberid,
        'ภาวะเสี่ยง': 'เสี่ยงโรคเบาหวาน (FINDRISC)',
        'ระดับความเสี่ยง': findrisc_level,
        'คะแนน': findrisc_raw_score # Using raw score as 'คะแนน'
    })

    # Calculate Cardiovascular Risk
    cardio_level, cardio_score, cardio_raw_score = calculate_cardiovascular_risk(age, gender, sys, waistline, height)
    disease_risk_list.append({
        'member_id': memberid,
        'ภาวะเสี่ยง': 'เสี่ยงโรคหัวใจและหลอดเลือด',
        'ระดับความเสี่ยง': cardio_level,
        'คะแนน': cardio_raw_score # Using raw score as 'คะแนน'
    })


disease_risk_wh = pd.DataFrame(disease_risk_list)

print(f"✅ Disease Risk Warehouse: {len(disease_risk_wh)} แถว")
display(disease_risk_wh.head())
```
