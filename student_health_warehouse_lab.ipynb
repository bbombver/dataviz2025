```python
# Step 2: ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à
# Hint: ‡πÉ‡∏ä‡πâ .groupby('memberid').size().reset_index(name='‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à')

screening_count = ncds_record.groupby('memberid').size().reset_index(name='‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à')

print(f"‚úì Screening count: {len(screening_count)} ‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å")
screening_count.head()
```

```python
# Step 3: JOIN ‡∏Å‡∏±‡∏ö campaign
# Hint: ‡πÉ‡∏ä‡πâ pd.merge() ‡πÇ‡∏î‡∏¢ left_on='campaignId', right_on='id'

campaign['district_code'] = campaign['district_code'].astype(str)

profile_campaign = pd.merge(latest_profiles, campaign[['id', 'district_code']], left_on='campaignId', right_on='id', how='left')

print(f"‚úì Profile + Campaign: {len(profile_campaign)} ‡πÅ‡∏ñ‡∏ß")
```

```python
# Step 5: JOIN ‡∏Å‡∏±‡∏ö record ‡πÅ‡∏•‡∏∞ screening_count
# Hint: JOIN 2 ‡∏Ñ‡∏£‡∏±‡πâ‡∏á - ‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÅ‡∏£‡∏Å‡∏Å‡∏±‡∏ö latest_records, ‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á‡∏Å‡∏±‡∏ö screening_count

profile_record = pd.merge(profile_location, latest_records, on='memberid', how='left', suffixes=('_profile', '_record'))

final_data = pd.merge(profile_record, screening_count, on='memberid', how='left')

# ‡πÅ‡∏õ‡∏•‡∏á NaN ‡πÉ‡∏ô‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à‡πÄ‡∏õ‡πá‡∏ô 0
final_data['‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à'] = final_data['‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à'].fillna(0).astype(int)

print(f"‚úì Final data: {len(final_data)} ‡πÅ‡∏ñ‡∏ß")
display(final_data.head())
```

```python
# Step 6: ‡∏™‡∏£‡πâ‡∏≤‡∏á Helper Functions
# TODO: ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤‡∏ï‡πà‡∏≤‡∏á‡πÜ

def get_gender_label(gender):
    """‡πÅ‡∏õ‡∏•‡∏á‡πÄ‡∏û‡∏®: 0‚Üí‡∏´‡∏ç‡∏¥‡∏á, 1‚Üí‡∏ä‡∏≤‡∏¢, ‡∏≠‡∏∑‡πà‡∏ô‡πÜ‚Üí999"""
    if gender == 0:
        return '‡∏´‡∏ç‡∏¥‡∏á'
    elif gender == 1:
        return '‡∏ä‡∏≤‡∏¢'
    else:
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'

def get_age_group(age):
    """
    ‡πÅ‡∏õ‡∏•‡∏á‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏≠‡∏≤‡∏¢‡∏∏:
    - < 20: '‡∏≠‡∏≤‡∏¢‡∏∏‡∏ï‡πà‡∏≥‡∏Å‡∏ß‡πà‡∏≤ 20 ‡∏õ‡∏µ'
    - 20-34: '‡∏≠‡∏≤‡∏¢‡∏∏‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 20-34 ‡∏õ‡∏µ'
    - 35-59: '‡∏≠‡∏≤‡∏¢‡∏∏‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 35-59 ‡∏õ‡∏µ'
    - >= 60: '‡∏≠‡∏≤‡∏¢‡∏∏ 60 ‡∏õ‡∏µ‡∏Ç‡∏∂‡πâ‡∏ô‡πÑ‡∏õ'
    """
    if age < 20:
        return '‡∏≠‡∏≤‡∏¢‡∏∏‡∏ï‡πà‡∏≥‡∏Å‡∏ß‡πà‡∏≤ 20 ‡∏õ‡∏µ'
    elif 20 <= age <= 34:
        return '‡∏≠‡∏≤‡∏¢‡∏∏‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 20-34 ‡∏õ‡∏µ'
    elif 35 <= age <= 59:
        return '‡∏≠‡∏≤‡∏¢‡∏∏‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 35-59 ‡∏õ‡∏µ'
    elif age >= 60:
        return '‡∏≠‡∏≤‡∏¢‡∏∏ 60 ‡∏õ‡∏µ‡∏Ç‡∏∂‡πâ‡∏ô‡πÑ‡∏õ'
    else:
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'


def convert_illness_history(value):
    """
    ‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢:
    - -1: '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
    - 1: '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢‡∏£‡∏ß‡∏°'
    - ‡∏≠‡∏∑‡πà‡∏ô‡πÜ: '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'
    """
    if value == -1:
        return '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
    elif value == 1:
        return '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢‡∏£‡∏ß‡∏°'
    else:
        return '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'

# TODO: ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏ï‡∏¥‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö drug_allergy, food_allergy, relatives_ncds
# (‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô convert_illness_history)

def convert_drug_allergy(value):
     """
     ‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏¢‡∏≤:
     - -1: '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
     - 1: '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏¢‡∏≤'
     - ‡∏≠‡∏∑‡πà‡∏ô‡πÜ: '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'
     """
     if value == -1:
         return '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
     elif value == 1:
         return '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏¢‡∏≤'
     else:
         return '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'

def convert_food_allergy(value):
     """
     ‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏≠‡∏≤‡∏´‡∏≤‡∏£:
     - -1: '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
     - 1: '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏≠‡∏≤‡∏´‡∏≤‡∏£'
     - ‡∏≠‡∏∑‡πà‡∏ô‡πÜ: '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'
     """
     if value == -1:
         return '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
     elif value == 1:
         return '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏≠‡∏≤‡∏´‡∏≤‡∏£'
     else:
         return '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'

def convert_relatives_ncds(value):
     """
     ‡πÅ‡∏õ‡∏•‡∏á‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏ç‡∏≤‡∏ï‡∏¥‡∏™‡∏≤‡∏¢‡∏ï‡∏£‡∏áNCDs:
     - -1: '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
     - 1: '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏ç‡∏≤‡∏ï‡∏¥‡∏™‡∏≤‡∏¢‡∏ï‡∏£‡∏áNCDs'
     - ‡∏≠‡∏∑‡πà‡∏ô‡πÜ: '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'
     """
     if value == -1:
         return '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
     elif value == 1:
         return '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏ç‡∏≤‡∏ï‡∏¥‡∏™‡∏≤‡∏¢‡∏ï‡∏£‡∏áNCDs'
     else:
         return '‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö'
```

```python
# ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏Ç‡∏≠‡∏á‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å
latest_profiles = ncds_profile.sort_values('dateadd').groupby('memberid').tail(1)
latest_records = ncds_record.sort_values('dateadd').groupby('memberid').tail(1)
screening_count = ncds_record.groupby('memberid').size().reset_index(name='‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à')

# JOIN ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
campaign['district_code'] = campaign['district_code'].astype(str)
profile_campaign = pd.merge(latest_profiles, campaign[['id', 'district_code']],
                            left_on='campaignId', right_on='id', how='left')
profile_location = pd.merge(profile_campaign, districts_df, on='district_code', how='left')
profile_record = pd.merge(profile_location,
                         latest_records[['memberid', 'drug_allergy', 'food_allergy', 'record_ncds', 'relatives_ncds']],
                         on='memberid', how='left')
final_data = pd.merge(profile_record, screening_count, on='memberid', how='left')
final_data['‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à'] = final_data['‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à'].fillna(0).astype(int)

# Helper functions
def handle_text_data(value):
    return 999 if pd.isna(value) or value == '' else str(value)

def handle_numeric_data(value):
    return 999 if pd.isna(value) else float(value)

def get_gender_label(gender):
    if pd.isna(gender): return 999
    return "‡∏´‡∏ç‡∏¥‡∏á" if int(gender) == 0 else "‡∏ä‡∏≤‡∏¢"

def get_age_group(age):
    if pd.isna(age) or age == 999 or age == 0: return 999
    if age < 20: return '‡∏≠‡∏≤‡∏¢‡∏∏‡∏ï‡πà‡∏≥‡∏Å‡∏ß‡πà‡∏≤ 20 ‡∏õ‡∏µ'
    if age < 35: return '‡∏≠‡∏≤‡∏¢‡∏∏‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 20-34 ‡∏õ‡∏µ'
    if age < 60: return '‡∏≠‡∏≤‡∏¢‡∏∏‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á 35-59 ‡∏õ‡∏µ'
    return '‡∏≠‡∏≤‡∏¢‡∏∏ 60 ‡∏õ‡∏µ‡∏Ç‡∏∂‡πâ‡∏ô‡πÑ‡∏õ'

def convert_value(value, labels):
    """‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤ -1, 1 ‡πÄ‡∏õ‡πá‡∏ô label"""
    if pd.isna(value) or value == '' or value == 999: return labels[0]
    if value == -1: return labels[1]
    if value == 1: return labels[2]
    return labels[0]

# ‡∏™‡∏£‡πâ‡∏≤‡∏á Warehouse
health_profile_wh = pd.DataFrame({
    'member_id': final_data['memberid'],
    '‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à': final_data['‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡∏Å‡∏≤‡∏£‡∏ï‡∏£‡∏ß‡∏à'],
    '‡πÄ‡∏û‡∏®': final_data['gender'].apply(get_gender_label),
    '‡∏≠‡∏≤‡∏¢‡∏∏': final_data['age'].apply(handle_numeric_data),
    '‡∏ï‡∏≥‡∏ö‡∏•_id': final_data['district_code'].apply(handle_text_data),
    '‡∏ï‡∏≥‡∏ö‡∏•': final_data['district_name'].apply(handle_text_data),
    '‡∏≠‡∏≥‡πÄ‡∏†‡∏≠_id': final_data['amphur_code'].apply(handle_numeric_data),
    '‡∏≠‡∏≥‡πÄ‡∏†‡∏≠': final_data['amphur_name'].apply(handle_text_data),
    '‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î_id': final_data['province_code'].apply(handle_numeric_data),
    '‡∏à‡∏±‡∏á‡∏´‡∏ß‡∏±‡∏î': final_data['province_name'].apply(handle_text_data),
    '‡πÄ‡∏Ç‡∏ï': final_data['zone'].apply(handle_numeric_data),
    '‡∏Å‡∏•‡∏∏‡πà‡∏°‡∏≠‡∏≤‡∏¢‡∏∏': final_data['age'].apply(get_age_group),
    '‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢_code': final_data['record_ncds'].apply(handle_numeric_data),
    '‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢': final_data['record_ncds'].apply(lambda x: convert_value(x, ['‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö', '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥', '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢‡∏£‡∏ß‡∏°'])),
    '‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏¢‡∏≤_code': final_data['drug_allergy'].apply(handle_numeric_data),
    '‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏¢‡∏≤': final_data['drug_allergy'].apply(lambda x: convert_value(x, ['‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö', '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥', '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÅ‡∏û‡πâ‡∏¢‡∏≤'])),
    '‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏≠‡∏≤‡∏´‡∏≤‡∏£_code': final_data['food_allergy'].apply(handle_numeric_data),
    '‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡πÅ‡∏û‡πâ‡∏≠‡∏≤‡∏´‡∏≤‡∏£': final_data['food_allergy'].apply(lambda x: convert_value(x, ['‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö', '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥', '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÅ‡∏û‡πâ‡∏≠‡∏≤‡∏´‡∏≤‡∏£'])),
    '‡∏ç‡∏≤‡∏ï‡∏¥‡∏™‡∏≤‡∏¢‡∏ï‡∏£‡∏áNCDs_code': final_data['relatives_ncds'].apply(handle_numeric_data),
    '‡∏ç‡∏≤‡∏ï‡∏¥‡∏™‡∏≤‡∏¢‡∏ï‡∏£‡∏áNCDs': final_data['relatives_ncds'].apply(lambda x: convert_value(x, ['‡πÑ‡∏°‡πà‡∏ó‡∏£‡∏≤‡∏ö', '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥', '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥‡∏Å‡∏≤‡∏£‡πÄ‡∏à‡πá‡∏ö‡∏õ‡πà‡∏ß‡∏¢‡∏£‡∏ß‡∏°']))
})

print(f"‚úÖ Health Profile Warehouse: {len(health_profile_wh)} ‡πÅ‡∏ñ‡∏ß")
health_profile_wh.head()
```

```python
# TODO: ‡∏™‡∏£‡πâ‡∏≤‡∏á NCDs Disease Warehouse
# Hint: ‡∏î‡∏π‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏à‡∏≤‡∏Å‡πÑ‡∏ü‡∏•‡πå detailed_health_warehouse_tutorial.ipynb

# ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏à‡∏≤‡∏Å ncds_record
latest_records = ncds_record.sort_values('dateadd').groupby('memberid').tail(1)

# ‡πÇ‡∏£‡∏Ñ‡∏Ç‡∏≠‡∏á‡∏ú‡∏π‡πâ‡∏ï‡∏£‡∏ß‡∏à
patient_diseases = latest_records[['memberid', 'record_cancer', 'record_chronic_kidney_disease',
                                   'record_cirrhosis', 'record_dementia', 'record_diabetes',
                                   'record_hyperlipidemia', 'record_hypertension', 'record_stroke']].copy()

patient_diseases_melted = patient_diseases.melt(id_vars='memberid', var_name='disease_code', value_name='status_code')
patient_diseases_melted['disease_type'] = '‡∏ú‡∏π‡πâ‡∏ï‡∏£‡∏ß‡∏à'

# ‡πÇ‡∏£‡∏Ñ‡∏Ç‡∏≠‡∏á‡∏ç‡∏≤‡∏ï‡∏¥
relatives_diseases = latest_records[['memberid', 'relatives_cancer', 'relatives_cirrhosis',
                                      'relatives_coronary_artery_disease', 'relatives_dementia',
                                      'relatives_diabetes', 'relatives_hyperlipidemia',
                                      'relatives_hypertension', 'relatives_stroke', 'relatives_obesity']].copy()

relatives_diseases_melted = relatives_diseases.melt(id_vars='memberid', var_name='disease_code', value_name='status_code')
relatives_diseases_melted['disease_type'] = '‡∏ç‡∏≤‡∏ï‡∏¥'

# ‡∏£‡∏ß‡∏°‡∏ï‡∏≤‡∏£‡∏≤‡∏á
ncds_disease_wh = pd.concat([patient_diseases_melted, relatives_diseases_melted])

# ‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤ status_code ‡πÄ‡∏õ‡πá‡∏ô label
def get_disease_status_label(status_code):
    if status_code == 1:
        return '‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
    elif status_code == 0:
        return '‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏õ‡∏£‡∏∞‡∏ß‡∏±‡∏ï‡∏¥'
    else:
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏' # ‡∏´‡∏£‡∏∑‡∏≠‡∏Ñ‡πà‡∏≤‡∏≠‡∏∑‡πà‡∏ô‡πÜ ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏°‡∏≤‡∏∞‡∏™‡∏°‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö -1 ‡∏´‡∏£‡∏∑‡∏≠ NaN

ncds_disease_wh['status_label'] = ncds_disease_wh['status_code'].apply(get_disease_status_label)

# ‡πÅ‡∏õ‡∏•‡∏á disease_code ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏ä‡∏∑‡πà‡∏≠‡πÇ‡∏£‡∏Ñ‡∏ó‡∏µ‡πà‡∏≠‡πà‡∏≤‡∏ô‡∏á‡πà‡∏≤‡∏¢
disease_code_map = {
    'record_cancer': '‡∏°‡∏∞‡πÄ‡∏£‡πá‡∏á',
    'record_chronic_kidney_disease': '‡πÑ‡∏ï‡πÄ‡∏£‡∏∑‡πâ‡∏≠‡∏£‡∏±‡∏á',
    'record_cirrhosis': '‡∏ï‡∏±‡∏ö‡πÅ‡∏Ç‡πá‡∏á',
    'record_dementia': '‡∏™‡∏°‡∏≠‡∏á‡πÄ‡∏™‡∏∑‡πà‡∏≠‡∏°',
    'record_diabetes': '‡πÄ‡∏ö‡∏≤‡∏´‡∏ß‡∏≤‡∏ô',
    'record_hyperlipidemia': '‡πÑ‡∏Ç‡∏°‡∏±‡∏ô‡πÉ‡∏ô‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏™‡∏π‡∏á',
    'record_hypertension': '‡∏Ñ‡∏ß‡∏≤‡∏°‡∏î‡∏±‡∏ô‡πÇ‡∏•‡∏´‡∏¥‡∏ï‡∏™‡∏π‡∏á',
    'record_stroke': '‡∏´‡∏•‡∏≠‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏™‡∏°‡∏≠‡∏á',
    'relatives_cancer': '‡∏°‡∏∞‡πÄ‡∏£‡πá‡∏á',
    'relatives_cirrhosis': '‡∏ï‡∏±‡∏ö‡πÅ‡∏Ç‡πá‡∏á',
    'relatives_coronary_artery_disease': '‡∏´‡∏•‡∏≠‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏´‡∏±‡∏ß‡πÉ‡∏à',
    'relatives_dementia': '‡∏™‡∏°‡∏≠‡∏á‡πÄ‡∏™‡∏∑‡πà‡∏≠‡∏°',
    'relatives_diabetes': '‡πÄ‡∏ö‡∏≤‡∏´‡∏ß‡∏≤‡∏ô',
    'relatives_hyperlipidemia': '‡πÑ‡∏Ç‡∏°‡∏±‡∏ô‡πÉ‡∏ô‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏™‡∏π‡∏á',
    'relatives_hypertension': '‡∏Ñ‡∏ß‡∏≤‡∏°‡∏î‡∏±‡∏ô‡πÇ‡∏•‡∏´‡∏¥‡∏ï‡∏™‡∏π‡∏á',
    'relatives_stroke': '‡∏´‡∏•‡∏≠‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏î‡∏™‡∏°‡∏≠‡∏á',
    'relatives_obesity': '‡∏≠‡πâ‡∏ß‡∏ô'
}
ncds_disease_wh['disease_name'] = ncds_disease_wh['disease_code'].map(disease_code_map)


# ‡∏à‡∏±‡∏î‡πÄ‡∏£‡∏µ‡∏¢‡∏á‡∏Ñ‡∏≠‡∏•‡∏±‡∏°‡∏ô‡πå
ncds_disease_wh = ncds_disease_wh[['memberid', 'disease_type', 'disease_code', 'disease_name', 'status_code', 'status_label']]

print(f"‚úÖ NCDs Disease Warehouse: {len(ncds_disease_wh)} ‡πÅ‡∏ñ‡∏ß")
display(ncds_disease_wh.head())
```

```python
# TODO: ‡∏™‡∏£‡πâ‡∏≤‡∏á Eating Behavior Warehouse
# Hint: ‡∏ï‡πâ‡∏≠‡∏á‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°

# ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏à‡∏≤‡∏Å ncds_habit
latest_habits = ncds_habit.sort_values('dateadd').groupby('memberid').tail(1).copy()

# Helper function ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°‡∏Å‡∏≤‡∏£‡∏Å‡∏¥‡∏ô
def convert_eating_behavior(value, mapping):
    if pd.isna(value):
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'
    return mapping.get(int(value), '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏')

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î mapping ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°
vegetable_map = {0: '‡πÑ‡∏°‡πà‡∏ó‡∏≥‡πÄ‡∏•‡∏¢', 1: '1-3 ‡∏ß‡∏±‡∏ô', 2: '4-6 ‡∏ß‡∏±‡∏ô', 3: '‡∏ó‡∏∏‡∏Å‡∏ß‡∏±‡∏ô'}
dessert_map = {0: '‡∏ó‡∏∏‡∏Å‡∏ß‡∏±‡∏ô', 1: '4-6 ‡∏ß‡∏±‡∏ô', 2: '1-3 ‡∏ß‡∏±‡∏ô', 3: '‡πÑ‡∏°‡πà‡∏ó‡∏≥‡πÄ‡∏•‡∏¢'} # ‡∏Å‡∏•‡∏±‡∏ö‡∏î‡πâ‡∏≤‡∏ô


# ‡∏™‡∏£‡πâ‡∏≤‡∏á Eating Behavior Warehouse
eating_behavior_wh = pd.DataFrame({
    'member_id': latest_habits['memberid'],
    '‡∏Å‡∏¥‡∏ô‡∏ú‡∏±‡∏Å_code': latest_habits['vegetable'],
    '‡∏Å‡∏¥‡∏ô‡∏ú‡∏±‡∏Å': latest_habits['vegetable'].apply(lambda x: convert_eating_behavior(x, vegetable_map)),
    '‡∏Å‡∏¥‡∏ô‡∏Ç‡∏ô‡∏°‡∏´‡∏ß‡∏≤‡∏ô_code': latest_habits['dessert'],
    '‡∏Å‡∏¥‡∏ô‡∏Ç‡∏ô‡∏°‡∏´‡∏ß‡∏≤‡∏ô': latest_habits['dessert'].apply(lambda x: convert_eating_behavior(x, dessert_map)),
})


print(f"‚úÖ Eating Behavior Warehouse: {len(eating_behavior_wh)} ‡πÅ‡∏ñ‡∏ß")
display(eating_behavior_wh.head())
```

```python
# TODO: ‡∏™‡∏£‡πâ‡∏≤‡∏á Comprehensive Behavior Warehouse
# Hint: ‡∏Ñ‡∏•‡πâ‡∏≤‡∏¢‡∏Å‡∏±‡∏ö Eating Behavior ‡πÅ‡∏ï‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°‡∏î‡πâ‡∏≤‡∏ô‡∏≠‡∏∑‡πà‡∏ô

# ‡∏î‡∏∂‡∏á‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏•‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏à‡∏≤‡∏Å ncds_habit
latest_habits = ncds_habit.sort_values('dateadd').groupby('memberid').tail(1).copy()

# Helper function ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ
def convert_behavior(value, mapping):
    if pd.isna(value):
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'
    return mapping.get(int(value), '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏')

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î mapping ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏û‡∏§‡∏ï‡∏¥‡∏Å‡∏£‡∏£‡∏°
movement_map = {0: '‡∏ô‡πâ‡∏≠‡∏¢‡∏°‡∏≤‡∏Å', 1: '‡∏ô‡πâ‡∏≠‡∏¢', 2: '‡∏õ‡∏≤‡∏ô‡∏Å‡∏•‡∏≤‡∏á', 3: '‡∏°‡∏≤‡∏Å'}
smoking_map = {0: '‡πÑ‡∏°‡πà‡∏™‡∏π‡∏ö', 1: '‡πÄ‡∏Ñ‡∏¢‡∏™‡∏π‡∏ö', 2: '‡∏™‡∏π‡∏ö‡∏õ‡∏±‡∏à‡∏à‡∏∏‡∏ö‡∏±‡∏ô'}
sleep_map = {0: '< 4 ‡∏ä‡∏°.', 1: '4-6 ‡∏ä‡∏°.', 2: '6-8 ‡∏ä‡∏°.', 3: '> 8 ‡∏ä‡∏°.'}
q2_map = {0: '‡πÑ‡∏°‡πà‡∏°‡∏µ/‡∏ô‡πâ‡∏≠‡∏¢', 1: '‡∏°‡∏µ/‡∏°‡∏≤‡∏Å'}
alcohol_map = {0: '‡πÑ‡∏°‡πà‡∏î‡∏∑‡πà‡∏°', 1: '‡∏î‡∏∑‡πà‡∏°‡∏ö‡∏≤‡∏á‡∏Ñ‡∏£‡∏±‡πâ‡∏á', 2: '‡∏î‡∏∑‡πà‡∏°‡∏õ‡∏£‡∏∞‡∏à‡∏≥'}
exercise_map = {0: '‡πÑ‡∏°‡πà‡∏ó‡∏≥‡πÄ‡∏•‡∏¢', 1: '1-3 ‡∏ß‡∏±‡∏ô', 2: '4-6 ‡∏ß‡∏±‡∏ô', 3: '‡∏ó‡∏∏‡∏Å‡∏ß‡∏±‡∏ô'}


# ‡∏™‡∏£‡πâ‡∏≤‡∏á Comprehensive Behavior Warehouse
comprehensive_behavior_wh = pd.DataFrame({
    'member_id': latest_habits['memberid'],
    '‡∏Å‡∏≤‡∏£‡πÄ‡∏Ñ‡∏•‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏´‡∏ß_code': latest_habits['movement'],
    '‡∏Å‡∏≤‡∏£‡πÄ‡∏Ñ‡∏•‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏´‡∏ß': latest_habits['movement'].apply(lambda x: convert_behavior(x, movement_map)),
    '‡∏Å‡∏≤‡∏£‡∏™‡∏π‡∏ö‡∏ö‡∏∏‡∏´‡∏£‡∏µ‡πà_code': latest_habits['smoking'],
    '‡∏Å‡∏≤‡∏£‡∏™‡∏π‡∏ö‡∏ö‡∏∏‡∏´‡∏£‡∏µ‡πà': latest_habits['smoking'].apply(lambda x: convert_behavior(x, smoking_map)),
    '‡∏Å‡∏≤‡∏£‡∏ô‡∏≠‡∏ô‡∏´‡∏•‡∏±‡∏ö_code': latest_habits['sleep'],
    '‡∏Å‡∏≤‡∏£‡∏ô‡∏≠‡∏ô‡∏´‡∏•‡∏±‡∏ö': latest_habits['sleep'].apply(lambda x: convert_behavior(x, sleep_map)),
    '‡∏†‡∏≤‡∏ß‡∏∞‡∏ã‡∏∂‡∏°‡πÄ‡∏®‡∏£‡πâ‡∏≤_code': latest_habits['2q_1'],
    '‡∏†‡∏≤‡∏ß‡∏∞‡∏ã‡∏∂‡∏°‡πÄ‡∏®‡∏£‡πâ‡∏≤': latest_habits['2q_1'].apply(lambda x: convert_behavior(x, q2_map)),
    '‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏Ñ‡∏£‡∏µ‡∏¢‡∏î_code': latest_habits['2q_2'],
    '‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏Ñ‡∏£‡∏µ‡∏¢‡∏î': latest_habits['2q_2'].apply(lambda x: convert_behavior(x, q2_map)),
    '‡∏Å‡∏≤‡∏£‡∏î‡∏∑‡πà‡∏°‡πÅ‡∏≠‡∏•‡∏Å‡∏≠‡∏Æ‡∏≠‡∏•‡πå_code': latest_habits['alcohol'],
    '‡∏Å‡∏≤‡∏£‡∏î‡∏∑‡πà‡∏°‡πÅ‡∏≠‡∏•‡∏Å‡∏≠‡∏Æ‡∏≠‡∏•‡πå': latest_habits['alcohol'].apply(lambda x: convert_behavior(x, alcohol_map)),
    '‡∏Å‡∏≤‡∏£‡∏≠‡∏≠‡∏Å‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏¢_code': latest_habits['exercise'],
    '‡∏Å‡∏≤‡∏£‡∏≠‡∏≠‡∏Å‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏Å‡∏≤‡∏¢': latest_habits['exercise'].apply(lambda x: convert_behavior(x, exercise_map))
})


print(f"‚úÖ Comprehensive Behavior Warehouse: {len(comprehensive_behavior_wh)} ‡πÅ‡∏ñ‡∏ß")
display(comprehensive_behavior_wh.head())
```

```python
# TODO: ‡∏™‡∏£‡πâ‡∏≤‡∏á Disease Risk Warehouse
# Hint: 1. ‡∏î‡∏∂‡∏á‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å memberdata (sys/dias, bmi, waistline, height)
#       2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏ó‡∏±‡πâ‡∏á 3 ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó
#       3. Loop ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì
#       4. ‡∏™‡∏£‡πâ‡∏≤‡∏á DataFrame ‡∏à‡∏≤‡∏Å‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå

# 1. ‡∏î‡∏∂‡∏á‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å memberdata
member_measurements = memberdata[memberdata['name'].isin(['sys/dias', 'bmi', 'waistline', 'height'])].copy()

# Pivot table to get measurements for each member in columns
member_measurements_pivot = member_measurements.pivot_table(
    index='memberid',
    columns='name',
    values='value',
    aggfunc='first'
).reset_index()

# Split sys/dias into sys and dias
if 'sys/dias' in member_measurements_pivot.columns:
    sys_dias = member_measurements_pivot['sys/dias'].str.split('/', expand=True)
    member_measurements_pivot['sys'] = pd.to_numeric(sys_dias[0], errors='coerce')
    member_measurements_pivot['dias'] = pd.to_numeric(sys_dias[1], errors='coerce')
    member_measurements_pivot = member_measurements_pivot.drop(columns=['sys/dias'])

# Convert other columns to numeric
for col in ['bmi', 'waistline', 'height']:
    if col in member_measurements_pivot.columns:
        member_measurements_pivot[col] = pd.to_numeric(member_measurements_pivot[col], errors='coerce')

# Merge with latest profile data for age, gender, relatives_diabetes
risk_data = pd.merge(member_measurements_pivot, latest_profiles[['memberid', 'age', 'gender']], on='memberid', how='left')
risk_data = pd.merge(risk_data, latest_records[['memberid', 'relatives_diabetes']], on='memberid', how='left')


# 2. ‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏ó‡∏±‡πâ‡∏á 3 ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó

def calculate_hypertension_risk(sys, dias):
    """‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏Ñ‡∏ß‡∏≤‡∏°‡∏î‡∏±‡∏ô‡πÇ‡∏•‡∏´‡∏¥‡∏ï‡∏™‡∏π‡∏á"""
    if pd.isna(sys) or pd.isna(dias):
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏', 999
    if sys >= 140 or dias >= 90:
        return '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏™‡∏π‡∏á', 1
    elif sys >= 120 or dias >= 80:
         return '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏õ‡∏≤‡∏ô‡∏Å‡∏•‡∏≤‡∏á', 0.5
    else:
        return '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏ï‡πà‡∏≥', 0

def calculate_findrisc_score(age, bmi, waistline, sys, relatives_diabetes):
    """‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì FINDRISC score ‡πÅ‡∏•‡∏∞‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á"""
    score = 0
    if pd.isna(age) or pd.isna(bmi) or pd.isna(waistline) or pd.isna(sys) or pd.isna(relatives_diabetes):
        return '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏', 999, 999

    # Age
    if age >= 45 and age <= 54: score += 2
    elif age >= 55 and age <= 64: score += 3
    elif age >= 65: score += 4

    # BMI
    if bmi >= 25 and bmi < 30: score += 1
    elif bmi >= 30: score += 3

    # Waistline (using female/male thresholds) - Assuming female if gender is 0, male if 1
    # We need gender for this, which is in latest_profiles. Let's assume we have it.
    # For simplicity here, let's use a generic threshold if gender is not available or unclear.
    # In a real scenario, you'd need to merge gender information.
    # Let's use generic thresholds for now or assume gender is available in risk_data
    # if gender == 0 (‡∏´‡∏ç‡∏¥‡∏á) and waistline >= 88: score += 4
    # elif gender == 1 (‡∏ä‡∏≤‡∏¢) and waistline >= 102: score += 4
    # elif gender == 0 (‡∏´‡∏ç‡∏¥‡∏á) and waistline >= 80 and waistline < 88: score += 3
    # elif gender == 1 (‡∏ä‡∏≤‡∏¢) and waistline >= 94 and waistline < 102: score += 3

    # Let's use a simplified approach for waistline based on available data structure (no gender directly here)
    # If you have gender, you'd modify this part. Assuming gender is available in risk_data now.
    # Need to merge gender from latest_profiles
    # Merged risk_data with latest_profiles already includes gender

    # Waistline based on gender
    gender = risk_data.loc[risk_data['memberid'] == memberid, 'gender'].iloc[0] if memberid in risk_data['memberid'].values else None

    if not pd.isna(gender):
        if int(gender) == 0: # ‡∏´‡∏ç‡∏¥‡∏á
            if waistline >= 88: score += 4
            elif waistline >= 80: score += 3
        elif int(gender) == 1: # ‡∏ä‡∏≤‡∏¢
            if waistline >= 102: score += 4
            elif waistline >= 94: score += 3
    # If gender is not available or invalid, skip waistline points or use a default logic if applicable

    # Sys (using the higher value if sys/dias exists)
    if sys >= 140: score += 2

    # Relatives with diabetes (assuming 1 means yes, -1 means no, 0 or NaN means unknown)
    if relatives_diabetes == 1: score += 5

    # Determine risk level
    if score < 7:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏ï‡πà‡∏≥'
    elif score >= 7 and score <= 11:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏õ‡∏≤‡∏ô‡∏Å‡∏•‡∏≤‡∏á'
    elif score >= 12 and score <= 14:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏™‡∏π‡∏á'
    elif score >= 15 and score <= 20:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏™‡∏π‡∏á‡∏°‡∏≤‡∏Å'
    else: # score > 20 or invalid score
        risk_level = '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'

    return risk_level, score, score

def calculate_cardiovascular_risk(age, gender, sys, waistline, height):
    """‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡πÇ‡∏£‡∏Ñ‡∏´‡∏±‡∏ß‡πÉ‡∏à‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏≠‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏î"""
    # This is a simplified calculation. Real cardiovascular risk scores (like Framingham, ASCVD)
    # are more complex and require more variables (cholesterol, blood pressure treatment, etc.).
    # Based on the prompt, we'll use the provided variables: age, gender, sys, waist/height ratio.

    risk_score = 0
    risk_level = '‡πÑ‡∏°‡πà‡∏£‡∏∞‡∏ö‡∏∏'

    if pd.isna(age) or pd.isna(gender) or pd.isna(sys) or pd.isna(waistline) or pd.isna(height) or height == 0:
         return risk_level, 999, 999

    # Age factor (example: higher age, higher risk)
    risk_score += age * 0.1

    # Gender factor (example: male higher risk)
    if int(gender) == 1: risk_score += 5

    # Systolic Blood Pressure factor
    risk_score += sys * 0.1

    # Waist-to-Height Ratio
    waist_height_ratio = waistline / height if height > 0 else 0
    risk_score += waist_height_ratio * 20 # Example factor

    # Determine risk level based on score (example thresholds)
    if risk_score < 20:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏ï‡πà‡∏≥'
    elif risk_score < 40:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏õ‡∏≤‡∏ô‡∏Å‡∏•‡∏≤‡∏á'
    else:
        risk_level = '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡∏™‡∏π‡∏á'

    return risk_level, risk_score, risk_score


# 3. Loop ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å‡πÅ‡∏•‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì & 4. ‡∏™‡∏£‡πâ‡∏≤‡∏á DataFrame ‡∏à‡∏≤‡∏Å‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
disease_risk_list = []

# Ensure all necessary columns are in risk_data, handle potential missing columns gracefully
required_cols = ['memberid', 'age', 'gender', 'bmi', 'waistline', 'height', 'sys', 'dias', 'relatives_diabetes']
for col in required_cols:
    if col not in risk_data.columns:
        risk_data[col] = np.nan # Add missing columns with NaN

for index, row in risk_data.iterrows():
    memberid = row['memberid']
    age = row['age']
    gender = row['gender']
    bmi = row['bmi']
    waistline = row['waistline']
    height = row['height']
    sys = row['sys']
    dias = row['dias']
    relatives_diabetes = row['relatives_diabetes'] # This comes from latest_records



    # Calculate Hypertension Risk
    hypertension_level, hypertension_score = calculate_hypertension_risk(sys, dias)
    disease_risk_list.append({
        'member_id': memberid,
        '‡∏†‡∏≤‡∏ß‡∏∞‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á': '‡πÇ‡∏£‡∏Ñ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏î‡∏±‡∏ô‡πÇ‡∏•‡∏´‡∏¥‡∏ï‡∏™‡∏π‡∏á',
        '‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á': hypertension_level,
        '‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô': hypertension_score
    })

    # Calculate FINDRISC
    findrisc_level, findrisc_score, findrisc_raw_score = calculate_findrisc_score(age, bmi, waistline, sys, relatives_diabetes)
    disease_risk_list.append({
        'member_id': memberid,
        '‡∏†‡∏≤‡∏ß‡∏∞‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á': '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡πÇ‡∏£‡∏Ñ‡πÄ‡∏ö‡∏≤‡∏´‡∏ß‡∏≤‡∏ô (FINDRISC)',
        '‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á': findrisc_level,
        '‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô': findrisc_raw_score # Using raw score as '‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô'
    })

    # Calculate Cardiovascular Risk
    cardio_level, cardio_score, cardio_raw_score = calculate_cardiovascular_risk(age, gender, sys, waistline, height)
    disease_risk_list.append({
        'member_id': memberid,
        '‡∏†‡∏≤‡∏ß‡∏∞‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á': '‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á‡πÇ‡∏£‡∏Ñ‡∏´‡∏±‡∏ß‡πÉ‡∏à‡πÅ‡∏•‡∏∞‡∏´‡∏•‡∏≠‡∏î‡πÄ‡∏•‡∏∑‡∏≠‡∏î',
        '‡∏£‡∏∞‡∏î‡∏±‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏™‡∏µ‡πà‡∏¢‡∏á': cardio_level,
        '‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô': cardio_raw_score # Using raw score as '‡∏Ñ‡∏∞‡πÅ‡∏ô‡∏ô'
    })


disease_risk_wh = pd.DataFrame(disease_risk_list)

print(f"‚úÖ Disease Risk Warehouse: {len(disease_risk_wh)} ‡πÅ‡∏ñ‡∏ß")
display(disease_risk_wh.head())
```
