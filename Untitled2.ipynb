# pandas : pararallel coordinate

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize'] = (5,5) #ขนาดของ figure
pd.plotting.parallel_coordinates(iris_df, 'target')
#กราฟนี้ช่วยให้เรา เปรียบเทียบหลายคุณลักษณะพร้อมกัน และเห็นชัดว่าทั้ง 3 species มี pattern ต่างกัน → สามารถใช้แยกกลุ่มดอกไม้ได้
```

# color

## color map สีที่กำหนดมาแล้วมากลุ่มสีนี้เข้ากัน

```python
matplotlib.rcParams['figure.figsize'] = (15,58) #ขนาดของ figure
plt.imshow(iris_df.iloc[:,:],interpolation='nearest',cmap='rainbow')
```

# color code

```python
matplotlib.rcParams['figure.figsize'] = (15,5) #ขนาดของ figure
pd.plotting_co(iris_df.iloc[:,:],interpolation='nearest',color= ['red','green','blue'])
```
