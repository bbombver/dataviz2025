จากคำให้การใน interview.csv ทราบว่าฆาตกรมีกระเป๋า Get Fit Now Gym membership number ที่ขึ้นต้นด้วย '48Z'
และเป็นสมาชิกระดับ gold
 หาข้อมูลสมาชิก Get Fit Now

```python
suspect_member = get_fit_now_member[
    (get_fit_now_member['id'].str.startswith('48Z')) &
    (get_fit_now_member['membership_status'] == 'gold')
]
```

```python
# ตรวจสอบว่าสมาชิกคนนี้เช็คอินในวันที่ 15 มกราคม 2018
suspect_checkin = get_fit_now_check_in[
    (get_fit_now_check_in['membership_id'] == suspect_member.iloc[0]['id']) &
    (get_fit_now_check_in['check_in_date'] == 20180115)
]

print("ข้อมูลสมาชิก Get Fit Now ที่เป็นผู้ต้องสงสัย:")
print(suspect_member)
```

```python
#การระบุตัวตนฆาตกร (Killer)
# เบาะแสจากคำให้การพยานในไฟล์ interview.csv:
# - มีกระเป๋ายิมของ Get Fit Now Gym
# - หมายเลขสมาชิกขึ้นต้นด้วย '48Z'
# - เป็นสมาชิกระดับ 'gold'
# - เช็คอินในวันที่เกิดเหตุ (15 มกราคม 2018)

killer_gym_id_prefix = '48Z'
killer_membership_status = 'gold'
case_date = 20180115

# กรองหาสมาชิกที่ตรงกับเบาะแสจากไฟล์ get_fit_now_member.csv
suspect_members = get_fit_now_member[
    (get_fit_now_member['id'].str.startswith(killer_gym_id_prefix)) &
    (get_fit_now_member['membership_status'] == killer_membership_status)
]

# ตรวจสอบการเช็คอินของผู้ต้องสงสัยในวันที่เกิดเหตุ
killer_checkin = get_fit_now_check_in[
    (get_fit_now_check_in['membership_id'].isin(suspect_members['id'])) &
    (get_fit_now_check_in['check_in_date'] == case_date)
]



if not miranda_priestly_info.empty:
    print(f"**ผู้บงการ:** {miranda_priestly_info['name'].iloc[0]}")
    print("\nข้อมูลส่วนตัวผู้บงการ:")
    print(miranda_priestly_info.to_string(index=False))
else:
    print("Jeremy Bowers")

print("\n" + "-" * 50 + "\n")

if not killer_checkin.empty:
    # ดึงข้อมูลสมาชิกและข้อมูลส่วนตัวของฆาตกร
    killer_membership_id = killer_checkin['membership_id'].iloc[0]
    killer_person_id = suspect_members[suspect_members['id'] == killer_membership_id]['person_id'].iloc[0]
    killer_info = person[person['id'] == killer_person_id]

    print(f"🔪 **ฆาตกร:** {killer_info['name'].iloc[0]}")
    print("\nข้อมูลส่วนตัวฆาตกร:")
    print(killer_info.to_string(index=False))
else:
    print("Jeremy Bowers")
```
